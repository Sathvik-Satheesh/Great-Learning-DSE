**(6). There are two array x and y ,drop the items in `X` which already exist in `Y`.**

**(9). Print the ranks for the given numeric array `x`.**
 
   
    Intake:
    np.random.seed(10)
    x = np.random.randint(20, size=10)
    print(x)
    [ 9  4 15  0 17 16 17  8  9  0]
    
    
    Result:
    [4 2 6 0 8 7 9 3 5 1]
    

```python
np.random.seed(10)
x = np.random.randint(20,size=10)
print(x)
temp = x.argsort()
print(temp)
ranks1 = temp.argsort()
print()
print(ranks1)
```
